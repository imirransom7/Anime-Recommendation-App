A dictionary comprehension that takes the list of anime titles and inserts it as the key for the new dictionary from its indeex, and the genre as the value from it's index as well

Here I make a loop and a nested for loop to get the unique genres

```python
se = set({})
for value in anime_dict.values():
    for v in value.split(','):
        # Using strip here because there was whitespace in some strings which
        # made so there were techincally duplicates
        se.add(v.strip())
unique_genres = sorted(list(se))        
```

## Creating Genre Dictionary
Making the genre dictionary so that I can turn it into a dataframe and then into a csv file

### Genre Funciton
I am going to create a fucntion to help auotmate the process; the keys are the genres, and the values will be all the anime associated with the genere

I will then interate through a list to automate the process of making the genre dicitonary

```python
genre_dict = {}


def genre_function(name: str) -> list[str]:
    values_list = []
    for key, value in anime_dict.items():
        if name in value:
            values_list.append(key)
    return values_list


for ug in unique_genres:
    genre_function(ug)
    genre_dict[ug] = genre_function(ug)
genre_dict
```
